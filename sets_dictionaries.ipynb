```python
### SETS(mutable,unordered,no duplicates)
numbers={1,2,3,2,4,1}
print("set:",numbers)
numbers.add(4)
numbers.add(6)
numbers.add(10)
numbers.discard(2)
numbers.discard(1)
print("set after adding and discarding:",numbers)
```

```python
### set operations
evens={2,4,6,8}
print("union:",numbers| evens)
print("intersection:",numbers & evens)
```
