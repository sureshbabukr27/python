```python
## 2E PRINT NUMBERS FROM 1 TO 5 USING WHILE LOOP
i=1
while i<=5:
    print(i)
    i=i+1;
```

```python
i=1
while i<=5:
    print(i)
    i=i+1;
    if i==4:
        break
```

```python
## 2G PRINT THW VALUE FROM 1 TO 5(SKIP THE ITERATION AT 4 USING WHILE) 
i=1
while i<=5:
    i=i+1
    if i==4:
        continue;
        print(i)
```
